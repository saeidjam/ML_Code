```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
dataset = pd.read_csv('C:\\Users\\Saeid\\OneDrive\\Desktop\\BostonHousing.csv')
```

```python
dataset.head()
```

```python
import statsmodels.api as sm 
```

```python
target = pd.DataFrame(data=dataset.values[:, 13], columns=["medv"])
```

```python
target.head()
```

```python
x = dataset["lstat"]
y = target["medv"]
x = sm.add_constant(x)

model = sm.OLS(y ,x).fit()
predictions = model.predict(x)

model.summary()
```

```python
plt.scatter(x=x,y=y)

plt.plot(x=x,y=y, lw=4, c='red', label ='regression line')

plt.xlabel('lstat', fontsize = 20)
plt.ylabel('medv', fontsize = 20)
plt.show()
```
