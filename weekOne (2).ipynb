```python
import statsmodels.api as sm 
```

```python
target = pd.DataFrame(data=dataset.values[:, 13], columns=["medv"])
```

```python
target.head()
```

```python
x = dataset["lstat"]
y = target["medv"]
x = sm.add_constant(x)

model = sm.OLS(y ,x).fit()
predictions = model.predict(x)

model.summary()
```
